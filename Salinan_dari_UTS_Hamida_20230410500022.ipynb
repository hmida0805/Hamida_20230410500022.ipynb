```python
import pandas as pd
import re
import nltk
import stanza

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Download resource NLTK
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Download model bahasa Indonesia untuk lemmatization
stanza.download('id')
```

# Bagian Baru

```python
df = pd.read_csv('genap.csv', sep=';')
df
```

```python
# Hapus baris yang Data-nya kosong
df = df.dropna(subset=['Data']).reset_index(drop=True)

# Hapus baris metadata
pattern_metadata = r'^(Sumber:|Membership:|Download aplikasi:)'
df = df[~df['Data'].str.contains(pattern_metadata, case=False, na=False)].reset_index(drop=True)

# Gabungkan jadi satu teks
text_asli = ' '.join(df['Data'].astype(str))

print("Teks Asli:")
print(text_asli)
```

```python
#Cleaning

def text_cleaning(text):
    text = re.sub(r'http\S+|www\S+', ' ', text)     # hapus URL
    text = re.sub(r'\d+', ' ', text)                # hapus angka
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)        # hapus tanda baca/simbol
    text = re.sub(r'\s+', ' ', text).strip()        # hapus spasi berlebih
    return text

clean_text = text_cleaning(text_asli)

print("Hasil Text Cleaning:")
print(clean_text)
```

```python
#Tokenization

tokens = word_tokenize(clean_text)

print("Hasil Tokenization:")
print(tokens)
```

```python
#Normalization

normalization_dict = {
    'rp': 'rupiah',
    'yg': 'yang',
    'dgn': 'dengan',
    'utk': 'untuk',
    'dlm': 'dalam',
    'tdk': 'tidak',
    'dll': 'dan lain lain'
}

def normalize_tokens(tokens):
    normalized = []
    for token in tokens:
        token_lower = token.lower()
        if token_lower in normalization_dict:
            normalized.append(normalization_dict[token_lower])
        else:
            normalized.append(token)
    return normalized

normalized_tokens = normalize_tokens(tokens)

print("Hasil Normalization:")
print(normalized_tokens[:50])
```

```python
#Stopword Removal

stop_words = set(stopwords.words('indonesian'))

tokens_no_stopwords = [word for word in normalized_tokens if word.lower() not in stop_words]

print("Hasil Stopword Removal:")
print(tokens_no_stopwords[:50])
```

```python
#Case Folding

case_folded_tokens = [word.lower() for word in tokens_no_stopwords]

print("Hasil Case Folding:")
print(case_folded_tokens[:50])
```

```python
#Stemming

factory = StemmerFactory()
stemmer = factory.create_stemmer()

stemmed_tokens = [stemmer.stem(word) for word in case_folded_tokens]

print("Hasil Stemming:")
print(stemmed_tokens[:50])
```

```python
#Lemmatization

nlp = stanza.Pipeline('id', processors='tokenize,pos,lemma', use_gpu=False)

text_for_lemma = ' '.join(stemmed_tokens)
doc = nlp(text_for_lemma)

lemmatized_tokens = []
for sentence in doc.sentences:
    for word in sentence.words:
        lemmatized_tokens.append(word.lemma)

print("Hasil Lemmatization:")
print(lemmatized_tokens[:50])
```

```python
#Ringkasan semua hasil

hasil_preprocessing = {
    "Text Asli": text_asli,
    "Text Cleaning": clean_text,
    "Tokenization": tokens,
    "Normalization": normalized_tokens,
    "Stopword Removal": tokens_no_stopwords,
    "Case Folding": case_folded_tokens,
    "Stemming": stemmed_tokens,
    "Lemmatization": lemmatized_tokens
}

for tahap, hasil in hasil_preprocessing.items():
    print(f"\n===== {tahap} =====")
    print(hasil)
```

```python
ringkasan_df = pd.DataFrame({
    'Tahap': [
        'Text Asli',
        'Text Cleaning',
        'Tokenization',
        'Normalization',
        'Stopword Removal',
        'Case Folding',
        'Stemming',
        'Lemmatization'
    ],
    'Hasil': [
        text_asli,
        clean_text,
        str(tokens),
        str(normalized_tokens),
        str(tokens_no_stopwords),
        str(case_folded_tokens),
        str(stemmed_tokens),
        str(lemmatized_tokens)
    ]
})

ringkasan_df
```
